# HARP2 L1C projected image generator

This script utilizes the Python package `nasa-pace-data-reader`, requiring Python 3.10 or higher. To install the latest version, use the command `pip install nasa-pace-data-reader` While it's advisable to use a virtual environment for this package, it's not obligatory.

Once installed, you can execute the script to project L1C HARP2 RGB images onto PlateCarre and Orthographic projections.

## HARP2 L1C RGB image example

Load the package and file

```python
# load the reader package
from nasa_pace_data_reader import L1, plot

# Set the dpi for the plots
dpi_ = 160

# Path to the L1C file
fileName = '/Users/aputhukkudy/Downloads/PACE/04-14/PACE_HARP2.20240414T064337.L1C.5km.nc'

# Read the file
l1c = L1.L1C()
l1c_dict = l1c.read(fileName)

# Load the plot class (default instrument is HARP2)
plt_ = plot.Plot(l1c_dict)
```

## Plot the RGB in plate carre projection

```python
# plot RGB in default plate carree projection
plt_.projectedRGB(normFactor=300, saveFig=True, dpi=dpi_)
```

## Plot the RGB in orthographic projection

```python
# plot in orthographic projection
plt_.projectedRGB(proj='Orthographic', normFactor=300, saveFig=True, dpi=dpi_)
```

## Plot the veggie RGB in orthographic projection

```python
# Plot the veggie RGB (using the modified RGB bands,[RGB= 670, 870, 443 nm])
plt_.projectedRGB(proj='Orthographic', viewAngleIdx=[31, 73, 83],
                  scale=[0.85, 1.4, 1], # scale the RGB bands modified to show the veggie RGB
                  normFactor=300, 
                  saveFig=True, dpi=dpi_)
```

## Plot the veggie RGB in platte carre projection

```python
# Plot the veggie RGB (using the modified RGB bands,[RGB= 670, 870, 443 nm])
plt_.projectedRGB(viewAngleIdx=[31, 73, 83],
                  scale=[0.85, 1.4, 1], # scale the RGB bands modified to show the veggie RGB
                  normFactor=300, 
                  saveFig=True, dpi=dpi_)
```
